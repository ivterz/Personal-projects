# Predicting Pima Indians Diabetes Cases

In this project we will use [this dataset](https://www.kaggle.com/uciml/pima-indians-diabetes-database) originally compiled by the National Institute of Diabetes and Digestive and Kidney Diseases. The objective of the dataset is to diagnostically predict whether or not a patient has diabetes, based on certain diagnostic measurements included in the dataset. Several constraints were placed on the selection of these instances from a larger database. In particular, all patients here are **female, at least 21 years old, of [Pima Indian heritage](https://en.wikipedia.org/wiki/Pima_people)**.

The datasets consists of several medical predictor variables:

- `Pregnancies` - Number of times pregnant;
- `Glucose` - Plasma glucose concentration a 2 hours in an oral glucose tolerance test
- `BloodPressure` - Diastolic blood pressure (mm Hg)
- `SkinThickness` - Triceps skin fold thickness (mm)
- `Insulin` - 2-Hour serum insulin (mu U/ml)
- `BMI` - Body mass index (weight in kg/(height in m)^2)
- `DiabetesPedigreeFunction` - Diabetes pedigree function
- `Age` - Age (years)

and one target variable - `Outcome` - Class variable (0 or 1).

Our goal is to build a machine learning model to accurately predict whether or not the patients in the dataset have diabetes or not.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

plt.style.use('fivethirtyeight')

import missingno as msm

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, cross_val_predict, GridSearchCV
from sklearn.metrics import plot_confusion_matrix, classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.feature_selection import RFECV
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('diabetes.csv')

df.shape
```

```python
df.head()
```

```python
df.describe(include='all').T
```

We see that there are no rows with NaN values. However, for a few of the columns, the minimum value is `0` when in reality it should be above `0`:

- `Glucose` - according to [this WebMD article](https://www.webmd.com/diabetes/how-sugar-affects-diabetes), normal glucose levels are around 70-80 mg/DL.
- `BloodPressure` - normal blood pressure is mostly between 90/60 mmHg and 120/80 mmHg, according to the [U.S. National Library of Medicine](https://medlineplus.gov/ency/article/007278.htm)
- `SkinThickness` - for female adults, the normal Tricep Skin Thickness should be about 18mm
- `Insulin` - check [this article](https://emedicine.medscape.com/article/2089224-overview) for insulin levels
- `BMI` - measures body fat based on height and weight


```python
missing = df[(df['Glucose'] == 0) | (df['BMI'] == 0) | (df['BloodPressure'] == 0) | (df['SkinThickness'] == 0) | (df['Insulin'] == 0)]

print('Total rows with missing data: {}'.format(missing.shape[0]))
```

```python
missing_dict = {}

cols = ['Glucose', 'BMI', 'BloodPressure', 'SkinThickness', 'Insulin']

for x in cols:
    by_x = df[df[x] == 0]
    m = by_x[x].count()
    missing_dict[x] = m
    
missing_s = pd.Series(missing_dict)
missing_s
```

In the above two cells we saw that there are:

- a total of 376 rows in which either one of the measurements is 0. That is almost 50% of the dataset.
- the number of times each measurement is 0. `Insulin` is 0 a total of 374 times.

Let's treat those values as missing.

```python
for c in cols:
    df[c] = df[c].replace(0,np.nan)
```

```python
msm.bar(df)
plt.title('Missing Values per column', y=1.09)
plt.show()
```

Since the missing values in `Glucose` and `BMI` are relatively few, we can drop the rows with missing values for those two columns.

For the remaining columns - let's visualize the distribution of the values.

```python
for x in ['BloodPressure', 'SkinThickness', 'Insulin']:
    sns.distplot(df[x])
    plt.title('Distribution of values in the\n{} column'.format(x))
    plt.show()
```

The `BloodPressure` distribution looks symetrical. Because of that, we will fill in the missing values with the mean of the column.

As for `SkinThickness`, `Insulin` - their distribution is right skewed. Because of that, we will fill in the missing values with the medium of the column.

```python
for c in ['SkinThickness', 'Insulin']:
    df[c] = df[c].fillna(df[c].median())
    
df['BloodPressure'] = df['BloodPressure'].fillna(df['BloodPressure'].mean())
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
msm.bar(df)
plt.title('Missing Values per Column after clean up')
plt.show()
```

```python
df.describe().T
```

We lost 16 rows which is about 2% of the original dataset.

Before we choose the best model to predict the `Outcome`, we will use **RandomForestClassifier** to find out which columns provide the most useful information for our predictions.

```python
df = df.select_dtypes(include='number')
df = df.dropna(axis=1)
            
all_X = df.drop('Outcome', axis=1)
all_y = df['Outcome']
    
rfc = RandomForestClassifier(random_state=1)
selector = RFECV(rfc, cv=10)
selector.fit(all_X, all_y)
    
best_columns = all_X.columns[selector.support_]
print(list(best_columns))
```

## Hyper Parameter Optimization

Now that we know which columns to use when we build our model, let's find out which model would give us best results. We will choose between Logistic Regression, K Nearest Neighbors and Random Forest Classifier.

```python
all_X = df[best_columns]
all_y = df['Outcome']

models = [
     {
        'name': 'LogisticRegression',
        'estimator': LogisticRegression(),
        'hyperparameters': 
            {
            'solver': ["newton-cg", "lbfgs", "liblinear"]
            }
        },
        {
        'name': 'KNeighborsClassifier',
        'estimator': KNeighborsClassifier(),
        'hyperparameters': 
            {
            "n_neighbors": range(1,20,2),
            "weights": ["distance", "uniform"],
            "algorithm": ["ball_tree", "kd_tree", "brute"],
            "p": [1,2]
            } 
        },
        {
        'name': 'RandomForestClassifier',
        'estimator': RandomForestClassifier(),
        'hyperparameters': 
            {
            "n_estimators": [4, 6, 9],
            "criterion": ["entropy", "gini"],
            "max_depth": [2, 5, 10],
            "max_features": ["log2", "sqrt"],
            "min_samples_leaf": [1, 5, 8],
            "min_samples_split": [2, 3, 5]          
            }
        }
    ]  

for m in models:
    print(m['name'])
    print(12*'-')
    grid = GridSearchCV(m['estimator'], param_grid=m['hyperparameters'], cv=10)
    grid.fit(all_X, all_y)
    m['best_params'] = grid.best_params_
    m['best_score'] = grid.best_score_
    m['best_model'] = grid.best_estimator_
    print('Best score {:.2f}%:'.format(grid.best_score_*100))
    print('Best parameters:', grid.best_params_)
    print('\n')
```

It looks like Random Forest Classifier give out the best result.

## Random Forest Classifier

Before we train the model, let's split the dataset into train:test. We will use the 75:25% split.

```python
train, test = train_test_split(df, test_size=.25, random_state=1)
```

```python
train.shape
```

```python
test.shape
```

```python
train.head()
```

```python
test.head()
```

```python
train_X = train[best_columns]
train_y = train['Outcome']

test_X = test[best_columns]
test_y = test['Outcome']
```

```python
rf = RandomForestClassifier(criterion='entropy', max_depth=10, max_features='log2', min_samples_leaf=8, min_samples_split=3, n_estimators=9)
scores = cross_val_score(rf, train_X, train_y, cv=5)
accuracy = scores.mean()

print('Random Forest Classifier accuracy on the training set: {:.2f}%'.format(accuracy*100))
```

```python
X_pred = cross_val_predict(rf, train_X, train_y)

train_df = train.copy()
train_df['RF_predicted'] = X_pred
train_df['RF_predicted_correct'] = train_df['Outcome'] == train_df['RF_predicted']

train_df.head()
```

```python
rf.fit(train_X, train_y)
y_pred = rf.predict(test_X)
```

```python
test_df = test.copy()
test_df['RF_predicted'] = y_pred
test_df['RF_predicted_correct'] = test_df['Outcome'] == test_df['RF_predicted']

test_df.head()
```

```python
print('Random forest Classifier accuracy on the test dataset: {:.2f}%'.format(test_df['RF_predicted_correct'].mean()*100))
```

Our model achieved 76% accuracy on both datasets. While the score itself isn't that high, it's reassuring to see that the model is doing equally well on data it has not seen. This means that it is not overfitting.

Now, let's try and see how many of the predictions were True Possitives, False Possitives, True Negatives and False Negatives.

```python
plot_confusion_matrix(rf, test_X, test_y, normalize='true', display_labels=['Non-diabetic', 'Diabetic'], cmap='Greens')
plt.title('Confusion Matrix')
plt.show()
```

```python
print('Precision Report:')
print('-'*20)
print(classification_report(test_y,y_pred))
```
