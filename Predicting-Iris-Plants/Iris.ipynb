# Predicting Iris Flowers

Fisher's Iris flower [dataset](https://archive.ics.uci.edu/ml/datasets/Iris) is a multivariate data set introduced by the British statistician, eugenicist, and biologist [Ronald Fisher](https://en.wikipedia.org/wiki/Ronald_Fisher) in his 1936 paper *The use of multiple measurements in taxonomic problems as an example of linear discriminant analysis*. More information on the history of the dataset can be found on [Wikipedia](https://en.wikipedia.org/wiki/Iris_flower_data_set).

In this project, we will use the dataset to train a machine learning model to categorize Iris flowers in three categories:

- Iris setosa
- Iris virginica 
- Iris versicolor

The data set consists of 50 samples from each of three species of Iris. Each sample has four features - the length and the width of the sepals and petals, in centimeters.



## Read in the data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

plt.style.use('fivethirtyeight')

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.neighbors import NearestNeighbors
```

```python
header = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']

iris = pd.read_csv('iris.data', names=header)
```

```python
iris.head()
```

```python
iris.shape
```

```python
iris.describe(include='all')
```

```python
iris.info()
```

```python
iris['class'].value_counts()
```

There are no missing values in the dataset. No cleaning is needed.

The samples are evenly distributed between the three different types of Iris flowers.

The `class` column is our target column, the one which we will try to predict. However, currently it is not in numerical format. Let's convert the column in categorical datatype. Then, we will create a new column called `target` which will include the categorical codes for the `class` column.

```python
iris['class'] = pd.Categorical(iris['class'])
iris['target'] = iris['class'].cat.codes

iris.head()
```

## Visualize the data

Let's visualize the rest of the features in order to see if there an even better feature to be used.

```python
sns.pairplot(iris, hue='class',diag_kind="hist")
plt.show()
```

If we look at the `petal_length` - even with a naked eye the three different plants are easily destiguishable. Let's have a closer look.

```python
sns.scatterplot(x=iris['petal_length'], y=iris['class'], hue = iris['class'])
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.title('Petal Length per Class', y=1.05)
plt.show()
```

Looking at the above scatterplot, we can conclude that:

- Iris-setosa have petal length no higer than 2 and petal width lower than 1;
- With a few exceptions, Iris-setosa have petal length no higer than 5 and petal width lower than 2;
- With a few exceptions, Iris-virginica have petal length higer than 5 and petal width higher than 4.5;

It looks like `petal_length` would be a good feature to use for manual classifications.

It looks like the Iris-setosa flowers are easy to destinguish even manually based on any of the features.

## Manual classification

Let's write a function which will classify the flowers manually.

```python
def manual_classification(row):
    """Predicts the Iris species by petal length
    """
    if row < 2:
        return 'Iris-setosa'
    elif row > 2 and row <= 5:
        return 'Iris-versicolor'
    else:
        return 'Iris-virginica'
```

```python
iris['manual_prediction'] = iris['petal_length'].apply(manual_classification)

iris['manual_correct'] = iris['manual_prediction'] == iris['class']

sum(iris['manual_correct'])
```

Using only the `petal_length` feature, our manual prediction was correct 140 out of 150 times. This gives a 93.33% correct rate.

Let's visualize our manual predictions.

```python
fig, axes = plt.subplots(2,2, figsize=(15,15))
plt.title('Actual vs Manually Predicted Class')
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='class', ax=axes[0,0])
axes[0,0].set_title('Actual class')
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='manual_prediction', ax=axes[0,1])
axes[0,1].set_title('Manually predicted class')
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='manual_correct', ax=axes[1,0])
axes[1,0].set_title('Correctly predicted')
axes[1,1].set_visible(False)
plt.show()
```

## Logistic Regression

In order to be able to train and test our models, we need to split the dataset in a train and test datasets. We will use the 75:25 split.

```python
train, test = train_test_split(iris, test_size=.25)
```

```python
train.shape
```

```python
test.shape
```

```python
train.head()
```

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

X_train = train[features]
y_train = train['target']
X_test = test[features]
y_test = test['target']
```

```python
lr = LogisticRegression(max_iter=2000, C=2)
accuracies = cross_val_score(lr, X_train, y_train, cv=5)

print('The Logistic Regression model achieved an average of {:.2f}% accuracy on the train dataset.'.
      format(accuracies.mean() * 100))
```

```python
predictions = cross_val_predict(lr, X_train, y_train, cv=5)

train_df = train.copy()
train_df['LR_predicted'] = predictions
train_df['LR_predicted_correct'] = train_df['target'] == train_df['LR_predicted']

train_df.head()
```

```python
fig, axes = plt.subplots(2,2, figsize=(15,15))
plt.title('Actual vs Logistic Regression Predicted Class\non Train Dataset')
sns.scatterplot(data=train_df, x='petal_length', y='petal_width', hue='target', ax=axes[0,0])
axes[0,0].set_title('Actual class')
sns.scatterplot(data=train_df, x='petal_length', y='petal_width', hue='LR_predicted', ax=axes[0,1])
axes[0,1].set_title('Logistic Regression predicted class')
sns.scatterplot(data=train_df, x='petal_length', y='petal_width', hue='LR_predicted_correct', ax=axes[1,0])
axes[1,0].set_title('Correctly predicted')
axes[1,1].set_visible(False)
plt.show()
```

Finally, let's run the logistic regression on the test dataset.

```python
lr.fit(X_train, y_train)
predictions = lr.predict(X_test)

accuracy = y_test == predictions
percentage = accuracy.mean()*100

print('{} out of the 38 Test samples were predicted correctly.'.format(sum(accuracy)))
print('This is a {:.2f}% accuracy.'.format(percentage))
```

```python
test_df = test.copy()

test_df['LR_predicted'] = predictions
test_df['LR_predicted_correct'] = test_df['target'] == predictions

test_df.head()
```

```python
fig, axes = plt.subplots(2,2, figsize=(15,15))
plt.title('Actual vs Logistic Regression Predicted Class\non Test Dataset')
sns.scatterplot(data=test_df, x='petal_length', y='petal_width', hue='target', ax=axes[0,0])
axes[0,0].set_title('Actual class')
sns.scatterplot(data=test_df, x='petal_length', y='petal_width', hue='LR_predicted', ax=axes[0,1])
axes[0,1].set_title('Logistic Regression predicted class')
sns.scatterplot(data=test_df, x='petal_length', y='petal_width', hue='LR_predicted_correct', ax=axes[1,0])
axes[1,0].set_title('Correctly predicted')
axes[1,1].set_visible(False)
plt.show()
```
